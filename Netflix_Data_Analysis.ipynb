# Netflix Content Analysis

This project analyzes Netflix's dataset to uncover insights about content trends, genres, and global distribution using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,5)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

## Movies vs TV Shows

```python
import matplotlib.pyplot as plt

type_counts = df['type'].value_counts()

plt.figure(figsize=(6,4))
type_counts.plot(kind='bar')
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10,5))
top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Content")
plt.show()
```

```python
yearly_content = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))
yearly_content.plot(kind='line')
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

Netflix has more Movies than TV Shows
Drama and Comedy are the most common genres
USA produces the highest number of shows
Content growth increased rapidly after 2015

```python
type_counts = df['type'].value_counts()

plt.figure()
type_counts.plot(kind='bar')
plt.title("Distribution of Movies vs TV Shows on Netflix", fontsize=14)
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
top_genres = genres.value_counts().head(10)

plt.figure()
top_genres.sort_values().plot(kind='barh')
plt.title("Top 10 Genres on Netflix", fontsize=14)
plt.xlabel("Count")
plt.show()
```

```python
yearly_content = df['year_added'].value_counts().sort_index()

plt.figure()
yearly_content.plot(kind='line', marker='o')
plt.title("Netflix Content Growth Over Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid()
plt.show()
```

```python
rating_counts = df['rating'].value_counts().head(10)

plt.figure()
rating_counts.plot(kind='bar')
plt.title("Top Content Ratings on Netflix")
plt.xticks(rotation=45)
plt.show()
```

Netflix has more Movies than TV Shows
Drama and International genres dominate the platform
USA is the leading content producer
Content growth increased rapidly after 2015
TV-MA is the most common rating
